```python
n=pd.Series(np.random.normal(size=10))
n.plot()
```

```python
n.cumsum().plot(grid=True)
```

```python
v=pd.DataFrame({'normal':np.random.normal(size=10),
              'gamma':np.random.gamma(1,size=10),
              'poisson':np.random.poisson(size=10)})
print(v)
v.cumsum(0).plot()
```

```python
v.cumsum(0).plot(subplots=True)
```

```python
v.cumsum(0).plot(secondary_y='normal')
```

```python
fig,axes=plt.subplots(nrows=3,ncols=1,figsize=(4,12))
for i,var in enumerate(['normal','gamma','poisson']):
    v[var].cumsum(0).plot(ax=axes[i],title=var)
axes[0].set_ylabel('cumulative sum')
```

# Simple Line Plots

```python
fig=plt.figure()
a=plt.axes()

s=np.linspace(0,20,100)
print(s)
a.plot(s,np.sin(s))
```

```python
plt.plot(s,np.sin(s))
```

## Line Colors and Styles

```python
plt.plot(s,np.sin(s-0),color='red')
plt.plot(s,np.sin(s-1),color='y')
plt.plot(s,np.sin(s-2),color='0.55')
```

```python
plt.plot(s,np.sin(s-0),linestyle='-')
plt.plot(s,np.sin(s-1),linestyle=':')
plt.plot(s,np.sin(s-2),linestyle='-.')
```

```python
plt.plot(s,np.sin(s-0))

plt.xlim(-1,15)
plt.ylim(-2,2)
```

```python
plt.plot(s,np.cos(s))
plt.title("Cosine Curve")
plt.xlabel("X")
plt.ylabel("Cos(X)")
```

## Scatter Plots with ``plt.scatter``

A second, more powerful method of creating scatter plots is the ``plt.scatter`` function, which can be used very similarly to the ``plt.plot`` function:

```python
plt.scatter(x,y,marker='o');
```

```python
rng=np.random.RandomState(10)
x=rng.randn(10)
y=rng.randn(10)
colors=rng.rand(10)
size=100*rng.rand(10)

plt.scatter(x,y,c=colors,s=size,alpha=0.5,cmap='Reds')
plt.colorbar();#show color scale
```

# Visualizing Errors

```python
x=np.linspace(0,2,10)
dy=0.1
y=np.sin(x)+dy*np.random.randn(10)
plt.errorbar(x,y,yerr=dy,fmt='.b');
```

## Histograms

A simple histogram can be a great first step in understanding a dataset.

```python
plt.style.use('seaborn-white')

d=np.random.randn(1000)
```

```python
plt.hist(d)
```

```python
plt.hist(d,bins=5, alpha=0.6,)
```

```python
x=np.random.normal(0,0.8,100)
y=np.random.normal(-2,1,100)

kwargs=dict(alpha=0.5,bins=10)
plt.hist(x,**kwargs)
plt.hist(y,**kwargs)
```

```python
counts,bin_edges=np.histogram(d,bins=5)
print(counts)
```

```python
mean=[5,2]
cov=[[3,1],[2,0.5]]
x,y = np.random.multivariate_normal(mean,cov,100)
```

```python
plt.hist2d(x, y, bins=10, cmap='Reds')
cb=plt.colorbar()
```

```python
plt.hexbin(x,y,gridsize=10,cmap='Reds')


cb=plt.colorbar()
```
